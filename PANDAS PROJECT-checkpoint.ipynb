```python
# extracting the records where a team won batting_first.
batting_first = df[df['win_by_runs']!= 0]
batting_first
```

```python
batting_first.head()
```

```python
# making a histogram.
plt.figure(figsize = (8,7))
plt.hist(batting_first['win_by_runs'],color = 'green')
plt.title('Distribution of runs')
plt.xlabel('runs')
plt.show()
```

```python
batting_first['winner'].value_counts()
```

```python
# finding out the numberof wins w.r.t each team after batting.
batting_first['winner'].value_counts()[0:3]
```

```python
batting_first['winner'].value_counts()[0:3].keys()
```

```python
batting_first['winner'].value_counts()[0:3].value_counts()
```

```python
# making a bar plot for top 3 teams with the most win after batting first
plt.figure(figsize = (7,7))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color = ['blue','green','red'])
plt.show()
```

```python
# Making a pie chart or bar graph for distribution of most wins after batting second.

plt.figure(figsize = (7,7))
plt.pie(list(batting_first['winner'].value_counts()[0:5]),labels=list(batting_first['winner']
                                                                      .value_counts()[0:5].keys()),autopct ='%0.1f%%',colors = ['blue','pink','orange','green','red'])
plt.show()
```

```python
a = batting_first['winner'].value_counts()[0:5].keys()
b = batting_first['winner'].value_counts()[0:5].value_counts()
                                         
                                         
```

```python
plt.figure(figsize = (7,7))
plt.pie(b,labels = a)
plt.pie(b,labels = b,autopct= '%0.1f%%',colors = ['yellow','grey','red','green','blue'])
plt.pie([1],colors =['w'],radius =0.4)
plt.show()
        
```
